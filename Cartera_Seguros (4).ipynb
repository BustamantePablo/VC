## CARPETA COMPLETA, DECIMALES

```python
# RUTA DE LA CARPETA QUE CONTIENE LOS ARCHIVOS
carpeta_seguros = "SEGUROS_AGOSTO"

### CREAR ARCHIVO EN EXCEL
with xlsxwriter.Workbook('CARTERA_AGOSTO_COMPANIASEGUROS.xlsx') as workbook: ### NOMBRE DEL ARCHIVO RESULTANTE
    worksheet = workbook.add_worksheet()

    header_format = workbook.add_format({
    'bold': True,
    'text_wrap': False,
    'valign': 'top',
    'fg_color': '#022596',  # Color RGB en hexadecimal
    'font_color': 'white',
    'border': 1
    })

    # ENCABEZADOS
    headers = ["COMPAÑIA SEGURO"] + list(dict_detalle.keys())
    for col, header in enumerate(headers):
        worksheet.write(0, col, header, header_format)

    row = 1  # FILA DONDE EMPIEZA A ESCRIBIR LOS DATOS

    # ITERA SOBRE TODOS LOS ARCGIVOS
    for nombre_archivo in os.listdir(carpeta_seguros):
        ruta_archivo = os.path.join(carpeta_seguros, nombre_archivo)

        # AUXILIAR
        if nombre_archivo.startswith('.') or os.path.isdir(ruta_archivo):
            continue

        try:
            # LEE EL CONTENIDO DE CADA ARCHIVO
            with open(ruta_archivo, "r") as file:
                lines = file.readlines()

                # PROCESA PRIMERA LINEA PARA VER EL NOMBRE DE LA COMPAÑIA (INICIO)
                lista_substrings_1 = []
                posicion_actual_1 = 0
                texto = lines[0]

                for registro, longitud in dict_nombre.items():
                    substring_1 = texto[posicion_actual_1:posicion_actual_1 + longitud]
                    lista_substrings_1.append(substring_1)
                    posicion_actual_1 += longitud

                List_nombre = []
                for i, (key, substring) in enumerate(zip(dict_nombre.keys(), lista_substrings_1)):
                    List_nombre.append(substring.strip())

                s1 = List_nombre[1] + "-" + List_nombre[2]
                s2 = List_nombre[3]
                s3 = List_nombre[4]

                sf = s2 + " " + s1 + ";" + s3
                # PROCESA PRIMERA LINEA PARA VER EL NOMBRE DE LA COMPAÑIA (END)

                # PROCESAR CADA LINEA DE DATOS EXCEPTO LA PRIMERA, PORQUE EL NOMBRE VA EN LA PRIMERA COLUMNA
                for line in lines[1:-1]:
                    line = line.strip()  # QUITA ESPACIOS DE ANTES Y AL FINAL
                    lista_substrings = []
                    posicion_actual = 0

                    # EXTRAE LOS SUB STRING BASADO EN EL DICCIONARIO
                    for longitud in dict_detalle.values():
                        substring = line[posicion_actual:posicion_actual + longitud]
                        lista_substrings.append(substring)
                        posicion_actual += longitud

                    # ESCRIBE SF EN LA PRIMERA COLUMNA (NOMBRE COMPAÑIA SEGURO)
                    worksheet.write(row, 0, sf)

                    # ESCRIBE LOS DEMÁS VALORES
                    for col, value in enumerate(lista_substrings, start=1):
                        columna_actual = headers[col]
                        try:
                            if columna_actual in dict_columnas_decimales:
                                if value.strip() == "":  # MANEJO DE CADENAS VACIAS
                                    valor_decimal = 0.0
                                else:
                                    valor_decimal = int(value) / 10**dict_columnas_decimales[columna_actual]
                                worksheet.write(row, col, valor_decimal)
                            else:
                                worksheet.write(row, col, value)
                        except ValueError as ve:
                            print(f"Error al procesar la columna {columna_actual} con valor '{value}' en {nombre_archivo}: {ve}")
                            worksheet.write(row, col, "ERROR")  # ERROR SI HA PROBLEMA CON EL VALOR

                    row += 1  # ESCRIBE SOBRE LA SIGUIENTE LINEA

        except Exception as e:
            print(f"Error al procesar {ruta_archivo}: {e}")

    # FORMATEAR COLUMNAS CON DECIMALES
    decimal_format = workbook.add_format({'num_format': '0.0000'})
    porcentaje_format = workbook.add_format({'num_format': '0.00'})

    for columna, decimales in dict_columnas_decimales.items():
        col_idx = headers.index(columna)
        worksheet.set_column(col_idx, col_idx, None, decimal_format if decimales == 4 else porcentaje_format)
```
